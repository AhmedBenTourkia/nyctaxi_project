```python
from modules.utils.date_utils import get_month_start_n_months_ago
from pyspark.sql.functions import date_format
```

```python
two_months_ago_start = get_month_start_n_months_ago(2)
```

```python
#Read the yellow_trips_enriched table and filter only to include trips with a pickup datetime

df = spark.read.table("nyctaxi.`02_silver`.yellow_trips_enriched").filter
(f"tpep_pickup_datetime > '{two_months_ago_start}'")
```

```python
df = df.withColumn("year_month", date_format("tpep_pickup_datetime", "yyyy-MM"))
```

```python
df.write.\
    option("path","abfss://nyctaxi-yellow@nyctaxistorage100.dfs.core.windows.net/yellow_trips_export").\
    mode("append").\
    format("json").\
    partitionBy("vendor","year_month").\
    saveAsTable("nyctaxi.04_export.yellow_trips_export")
```
